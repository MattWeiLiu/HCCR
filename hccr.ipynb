```python
import cv2
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_bbox(img_path, result, color=(255, 0, 0), thickness=2):
    img_path = img_path.copy()
    cv2.line(img_path, (result[0]['x'], result[0]['y']), (result[1]['x'], result[1]['y'],), color, thickness)
    cv2.line(img_path, (result[1]['x'], result[1]['y']), (result[2]['x'], result[2]['y'],), color, thickness)
    cv2.line(img_path, (result[2]['x'], result[2]['y']), (result[3]['x'], result[3]['y'],), color, thickness)
    cv2.line(img_path, (result[3]['x'], result[3]['y']), (result[0]['x'], result[0]['y'],), color, thickness)
    return img_path
```

```python
input_img = cv2.imread(' 2021-01-15 5.53.07.png')

plt.figure(figsize=(10,10))
plt.imshow(input_img)
plt.show()
```

```python

with open('test5_output.json', 'r', encoding='utf-8') as f:
    output = json.load(f)
    

blocks = output["responses"][0]["fullTextAnnotation"]["pages"][0]["blocks"]

img = input_img.copy()

for block in blocks:
    
    img = draw_bbox(img, block['boundingBox']["vertices"], color=(0, 0, 255))
    
    
    words = block["paragraphs"][0]["words"]
    for word in words:
        img = draw_bbox(img, word["boundingBox"]["vertices"])

plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```
